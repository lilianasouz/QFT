```python
from  qiskit import*
circuit = QuantumCircuit(7,7)
from math import pi
```

```python
circuit.h([0])
circuit.cp(pi,1,0)
circuit.cp(pi/2,2,0)
circuit.cp(pi/4,3,0)
circuit.cp(pi/8,4,0)
circuit.cp(pi/16,5,0)
circuit.barrier()

circuit.h([1])
circuit.cp(pi/2,2,1)
circuit.cp(pi/4,3,1)
circuit.cp(pi/8,4,1)
circuit.cp(pi/16,5,1)
circuit.barrier()

circuit.h([2])
circuit.cp(pi/4,3,2)
circuit.cp(pi/8,4,2)
circuit.cp(pi/16,5,2)
circuit.barrier()

circuit.h([3])
circuit.cp(pi/8,4,3)
circuit.cp(pi/16,5,3)
circuit.barrier()

circuit.h([4])
circuit.cp(pi/16,5,4)
circuit.barrier()

circuit.h([5])
circuit.barrier()
circuit.x([0])

circuit.draw(output = 'mpl')
```
